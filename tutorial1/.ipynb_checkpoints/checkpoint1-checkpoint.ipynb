<h2>Exploratory Data Analysis (EDA) </h2>

Chosen Outcome Variable for Test: *AVG_MATH_8_Score*

How many years of data are logged in our dataset? 

```python
unique_years = df['YEAR'].unique()
num_years = len(unique_years)
num_years
```

Let's compare Michigan to Ohio. Which state has the higher average outcome score across all years?

```python
michigan_data = df[df['STATE'] == 'MICHIGAN']
ohio_data = df[df['STATE'] == 'OHIO']

average_math_8_michigan = michigan_data['AVG_MATH_8_SCORE'].mean()
average_math_8_ohio = ohio_data['AVG_MATH_8_SCORE'].mean()

if average_math_8_michigan > average_math_8_ohio:
    result = "Michigan has a higher average 'avg_math_8' score across all years."
elif average_math_8_michigan < average_math_8_ohio:
    result = "Ohio has a higher average 'avg_math_8' score across all years."
else:
    result = "Michigan and Ohio have the same average 'avg_math_8' score across all years."

print(result)
```

```python
michigan_years = michigan_data['YEAR']
michigan_avg_math_8_scores = michigan_data['AVG_MATH_8_SCORE']

ohio_years = ohio_data['YEAR']
ohio_avg_math_8_scores = ohio_data['AVG_MATH_8_SCORE']

plt.figure(figsize=(12, 6))
plt.plot(michigan_years, michigan_avg_math_8_scores, marker='o', linestyle='-', label='Michigan')
plt.plot(ohio_years, ohio_avg_math_8_scores, marker='o', linestyle='-', label='Ohio')
plt.title('Comparison of 8th Grade Math Scores: Michigan vs. Ohio')
plt.xlabel('Year')
plt.ylabel('Average Math Score (8th Grade)')
plt.legend()
plt.grid(True)
plt.show()
```

Line plot that compares the average 8th-grade math scores of Michigan and Ohio over the years. We can see that Ohio has a higher average 8th grade math score over the years. 

Bar Plot that compares the average 8th-grade math scores for Michigan and Ohio

Find the average for your outcome score across all states in 2019

```python
math_8_2019 = df[df['YEAR'] == 2019]

average_math_8_2019 = math_8_2019['AVG_MATH_8_SCORE'].mean()

print("Average 8th-Grade Math Score for All States in 2019:", average_math_8_2019)
```

Find the maximum outcome score for every state. 

Refer to the `Grouping and Aggregating` section in Tutorial 0 if you are stuck.

```python
max_avg_math_8_score_by_state = df.groupby('STATE')['AVG_MATH_8_SCORE'].max().reset_index()

max_avg_math_8_score_by_state.columns = ['STATE', 'Max_Avg_Math_8_Score']

print(max_avg_math_8_score_by_state)
```

```python
plt.figure(figsize=(12, 8))
plt.bar(max_avg_math_8_score_by_state['STATE'], max_avg_math_8_score_by_state['Max_Avg_Math_8_Score'])
plt.xlabel('State')
plt.ylabel('Max Average Math 8th Grade Score')
plt.title('Maximum Average 8th Grade Math Scores by State')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.show()
```

This bar plot illustrates the maximum average 8th-grade math scores achieved by different states. by analyzing this graph, we can observe significant variations in educational performance across states, with some states consistently achieving higher math scores than others.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
df
```

```python
X = df[['ENROLL_8','AVG_MATH_4_SCORE','AVG_READING_8_SCORE']].dropna()
y = df.loc[X.index]['AVG_MATH_8_SCORE']
y.fillna(y.median(), inplace=True)
```

```python
# X_train, X_test, y_train, y_test = train_test_split(
#      X, y, test_size=, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
print(model.intercept_)
print(model.coef_)
```

```python
model.score(X_test, y_test)
```

```python
np.mean(model.predict(X_test)-y_test)
```

```python
np.mean(np.abs(model.predict(X_test)-y_test))
```

```python
np.mean((model.predict(X_test)-y_test)**2)**0.5
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'AVG_READING_8_SCORE'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('Math 8 score')
plt.title("Model Behavior On Training Set")
```

```python
col_name = 'AVG_READING_8_SCORE'

f = plt.figure(figsize=(12,6))
plt.scatter(X_test[col_name], y_test, color = "blue")
plt.scatter(X_test[col_name], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel('Math 8 score')
plt.title("Model Behavior on Testing Set")
```
